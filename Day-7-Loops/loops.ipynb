# Sunucuya Bağlanma Uygulaması

```python
import time
import random

MAKSIMUM_DENEME = 5
deneme_sayisi = 0

while deneme_sayisi < MAKSIMUM_DENEME:
    deneme_sayisi += 1
    print(f"{deneme_sayisi}. deneme: Sunucuya bağlanılıyor...")
    time.sleep(0.5)

    if random.choice([True, False, False, False]):
        print("✅ Bağlantı başarılı!")
        break
    if deneme_sayisi == 2:
        print("İkinci deneme başarısız oldu, hemen yeniden denenecek.")
        continue
    print("Bağlantı başarısız. Tekrar deneniyor...")
else:
    print("Tüm denemeler başarısız. Bağlantı kurulamadı")
```

```python
import time

baslangic = time.time()

sonuc = sum((x * x for x in range(1_000_000_000)))

bitis = time.time()

print("Toplam:", sonuc)
print("Çalışma süresi:", bitis - baslangic, "saniye")
```
