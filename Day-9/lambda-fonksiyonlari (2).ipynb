```python
def faktoriyel(n):
    print(f"faktoriyel({n}) çağrıldı.")
    if n==0:
        print("Base case'e ulaşıldı, 1 döndürüldü.")
        return 1
    else: 
        sonuc = n * faktoriyel(n-1)
        print(f"{n}* factoriyel({n-1}) = {sonuc}")
        return sonuc
```

```python
faktoriyel(4)
```

```python
def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1)+fib(n-2)
```

```python
fib(0)
```

```python
fib(1)
```

```python
fib(2) # fib(1)+fib(0) = 1 + 0 =1
```

```python
fib(3) #  fib(2)+fib(1) = fib(1)+fib(0)+fib(1) = 1 + 0 + 1 = 2
```

```python
def binary_search_recursive(liste, hedef, baslangıc, bitis):
    if baslangıc > bitis:
        return -1 # sayı bulunamadı

    orta = (baslangıc + bitis) //2 

    if liste[orta] == hedef:
        return orta
    elif liste[orta]<hedef:
        return binary_search_recursive(liste, hedef, orta+1, bitis)
    else:
        return binary_search_recursive(liste, hedef, baslangıc, orta-1)
```

```python
sayilar = [3, 8, 12, 17, 25, 32, 39, 41, 53, 60]
```

```python
indeks = binary_search_recursive(sayilar, 25, 0, len(sayilar)-1)

if indeks == -1:
    print(f"Sayı bulunmadı! ")
else:
    print(f"Sayı bulundu. indeksi: {indeks}")
```

```python
indeks = binary_search_recursive(sayilar, 53, 0, len(sayilar)-1)

if indeks != -1:
    print(f"Sayı bulunmadı! indeksi: {indeks}")
else:
    print("Sayı listede yok")
```
